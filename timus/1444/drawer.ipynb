```python
from matplotlib.pyplot import *
import subprocess

f = open('in.txt')
t = int(f.readline())
tests = []
for idx in range(t):
    n = int(f.readline())
    pt = []
    for num_line in range(n):
        line = f.readline()
        pt.append([int(i) for i in line.split()])
    tests.append(pt)
print(len(tests))
```

```python
out = open('out.txt')
ans = []
is_passed = []
for idx in range(t):
    #passed = out.readline()
    #is_passed.append("PASSED" in passed)
    res_n = int(out.readline())
    res = []
    for line_num in range(res_n):
        res.append(int(out.readline()))
    ans.append(res)
print(len(ans))
```

```python
for num in range(t):
    p = tests[num]
    res = ans[num]
    print(p)
    print(res)
    plot([v[0] for v in p],[v[1] for v in p],'ro')
    plot([p[i-1][0] for i in res], [p[i-1][1] for i in res],'b-')
    axis('equal'); grid(); show()
```
